```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import exposure
# Load the image
input_image = cv2.imread('/content/drive/MyDrive/Github/1.png', cv2.IMREAD_GRAYSCALE)
# Read the image in grayscale
# Create a CLAHE (Contrast Limited Adaptive Histogram Equalization) object
clahe = cv2.createCLAHE(clipLimit=0.03, tileGridSize=(8, 8))
# Apply AHE to the image
ahe_image = clahe.apply(input_image)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(input_image, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(ahe_image, cmap='gray')
plt.title('AHE Image')
plt.axis('off')
plt.tight_layout()
plt.show()
```
