```python
from PIL import Image
import glob
```

```python
image_list = []
resized_images = []
```

```python
for filename in glob.glob('C:/Users/DCL/Desktop/HMI/I/*.png'):
  print(filename)
  img = Image.open(filename)
  image_list.append(img)
```

```python
for image in image_list:
  image.show()
  image = image.resize((900,900))
  resized_images.append(image)
```

```python
for (i,new) in enumerate(resized_images):
    new.save('{}{}{}'.format("C:/UsersDCL\Desktop\History of ML\I",i+1,'.png'))
```
